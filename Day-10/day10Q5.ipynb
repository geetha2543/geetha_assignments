# E. Plotting normal random numbers and show that addition of two Gaussian is Gaussian and addition of all is Gaussian

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.random.normal(0, 1, 10000)
x2 = np.random.normal(0, 1, 10000)
```

```python
x=x1+x2
```

```python
n = 5
n_gaussians = np.zeros(10000)
for _ in range(n):
    n_gaussians += np.random.normal(0, 1, 10000)
```

```python
plt.figure(figsize=(16, 4))

plt.subplot(1, 4, 1)
plt.hist(x1, bins=50, density=True, alpha=0.6, label='x1')
plt.title('x1')

plt.subplot(1, 4, 2)
plt.hist(x2, bins=50, density=True, alpha=0.6, label='x2')
plt.title('x2')

plt.subplot(1, 4, 3)
plt.hist(x, bins=50, density=True, alpha=0.6, label='x')
plt.title('2 guassians')

plt.subplot(1, 4, 4)
plt.hist(n_gaussians, bins=50, density=True, alpha=0.6, label='n_gaussians')
plt.title('n gaussians ')
```
