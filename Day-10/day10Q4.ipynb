# Create two frequencies 5Hz and 50Hz sin signals 
# Draw FFT components 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f1=5
f2=50
fs=500
t = np.linspace(0,1,fs, endpoint=False)
```

```python
s1=np.sin(2*np.pi*f1*t)
s2=np.sin(2*np.pi*f2*t)
```

```python
fft1 = np.fft.fft(s1) 
fft_freq1 = np.fft.fftfreq(len(s1), 1/fs)
```

```python
fft2 = np.fft.fft(s2) 
fft_freq2 = np.fft.fftfreq(len(s2), 1/fs)
```

```python
df1 = pd.DataFrame({'Frequency': fft_freq1, 'Amplitude': np.abs(fft1)})
df2 = pd.DataFrame({'Frequency': fft_freq2, 'Amplitude': np.abs(fft2)})
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(df1['Frequency'], df1['Amplitude'])
plt.title("FFT of 5 Hz Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)


plt.subplot(1, 2, 2)
plt.plot(df2['Frequency'], df2['Amplitude'])
plt.title("FFT of 50 Hz Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.tight_layout()
plt.show()
```
