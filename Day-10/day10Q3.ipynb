<h2>2.How many % of matched won by our player? ('winner')</h2>

```python
total_matches=player.shape[0]
matches_won=player[player['winner']=='Roger Federer'].shape[0]
win_rate=(matches_won/total_matches)*100
win_rate
```

<h3>3.Proportion of double faults wrt total points in each match This number is an indicator of the player's state of mind, his level of self-confidence, his willingness to take risks while serving, and other parameters. columns: 'player1 double faults' and 'player1 total points total' Display simple stats of above</h3>

```python
player['Df_to_totalpoints']=player['player1 double faults']/player['player1 total points total']
player.Df_to_totalpoints
```

```python
player['Df_to_totalpoints'].describe()
```

<h2>4.Average Win per surface</h2>

```python
wins = player[player['winner'] == 'Roger Federer']
total_surfaces = player['surface'].value_counts()
win_surface = wins['surface'].value_counts()
win_surface_rate = (win_surface/total_surface)*100 
win_surface_rate
```

<h2>Display the proportion of double faults as a function of the tournament date, 'start date' Trend: display average double faults in each year</h2>

```python
player['start date']=pd.to_datetime(player['start date'],errors='coerce')
player['year']=player['start date'].dt.year
avgdoublefaults_to_year=player.groupby('year')['Df_to_totalpoints'].mean()
avgdoublefaults_to_year
```
