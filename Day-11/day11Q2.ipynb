# B. Use LabelEncoder and MinMaxScalar and Use pipeline 
# Then GridSearch on param_grid = dict(lr__C=[0.1, 1, 10]) using Logistic regression 

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
le=LabelEncoder()
y=le.fit_transform(df["diagnosis"])
```

```python
X = df.drop(columns=['diagnosis'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
min_max=MinMaxScaler()
lr=LogisticRegression()
pipe=Pipeline([("min_max",min_max),("lr",lr)])
```

```python
pipe
```

```python
param_grid = dict(lr__C=[0.1, 1, 10])
```

```python
grid_search = GridSearchCV(pipe, param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

```python
grid_search.score(X_train,y_train),grid_search.score(X_test,y_test)
```

```python
y_pred=grid_search.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```
