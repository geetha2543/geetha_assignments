```python
from tpot import TPOTClassifier
```

```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"C:\Users\Geetha\handson\day10\code\data\titanic_train.csv")
df.head()
```

```python
df = df.rename(columns={"survived": "class"})
df = df.drop(['name', 'ticket', 'cabin'], axis=1)
le = LabelEncoder()
df['sex'] = le.fit_transform(df['sex'])
```

```python
df.head()
```

```python
df['embarked'] = df['embarked'].map({'S': 0, 'C': 1, 'Q': 2})
df['embarked'] = df['embarked'].fillna(0)
```

```python
fill=df["age"].median()
df["age"]=df["age"].fillna(fill)
```

```python
y=df["class"]
x = df.drop('class', axis=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1)
cf = TPOTClassifier(generations=5, population_size=20, verbosity=2, random_state=42)
```

```python
cf.fit(x_train, y_train)
```

```python
cf.score(x_train, y_train),cf.score(x_test,y_test)
```

```python
cf.predict(x_test)
```

```python
cf.export("tpot_titanic_pipe.py")
```
